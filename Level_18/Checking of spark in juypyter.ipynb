```python
import os
import shutil
import time
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder \
    .appName("FileSizeComparison") \
    .getOrCreate()

# Load your dataset into DataFrame (assuming the path to your dataset is correct)
input_file_path = "ipl_commentary_data.csv"
df = spark.read.csv(input_file_path, header=True, inferSchema=True)

# Show the DataFrame to ensure it's loaded correctly
df.show()

# Define paths for CSV and Parquet outputs
# For Windows, you can use:
csv_output_path = r"C:\path\to\your\output\commentary_data_csv"
parquet_output_path = r"C:\path\to\your\output\commentary_data_parquet"

# Start timing
start_time = time.time()

# Write DataFrame to CSV and Parquet
df.write.mode("overwrite").csv(csv_output_path, header=True)
df.write.mode("overwrite").parquet(parquet_output_path)

# Log execution time for writing
print(f"Execution Time for writing: {time.time() - start_time} seconds")

# Function to get file size in MB
def get_file_size_in_mb(file_path):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(file_path):
        for filename in filenames:
            total_size += os.path.getsize(os.path.join(dirpath, filename))
    return total_size / (1024 * 1024)  # Convert bytes to MB

# Get file sizes for CSV and Parquet
csv_size = get_file_size_in_mb(csv_output_path)
parquet_size = get_file_size_in_mb(parquet_output_path)

# Print file sizes
print(f"CSV File Size: {csv_size:.2f} MB")
print(f"Parquet File Size: {parquet_size:.2f} MB")

# Download the CSV file from DBFS (for Databricks users)
# You can download the CSV from the DBFS path by using the DBFS CLI in Databricks or through the web UI
# For local paths, skip this step

# If using Databricks CLI or other tools:
# dbfs cp /dbfs/tmp/commentary_data_csv /dbfs/tmp/my_downloaded_csv.csv

# Optionally, you can delete temporary output files after comparing
shutil.rmtree(csv_output_path)
shutil.rmtree(parquet_output_path)

# Log execution time for file comparison
print(f"Total Execution Time: {time.time() - start_time} seconds")
```
