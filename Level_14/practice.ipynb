```python
import requests
from bs4 import BeautifulSoup

r = requests.get('https://www.cricbuzz.com/cricket-scorecard-archives')

soup = BeautifulSoup(r.content, 'html.parser')

s = soup.find('div', class_='entry-content')
content = soup.find_all('p')

print(content)
```
